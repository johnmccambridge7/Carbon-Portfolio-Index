```python
spy_text_data = []

with open(etf_basket_csv) as f:
    for i, line in enumerate(f.readlines()):
        if i > 13: # skip initial description of ETF

            stripped_line = line.rstrip('\n')[::-1]

            name, ticker, holding_percentage = stripped_line.split(",", 2)
            name, ticker, holding_percentage = name[::-1], ticker[::-1], holding_percentage[::-1]

            # print(name, ticker, holding_percentage)
            
            ticker = ticker.rstrip(' ')
            print(ticker)
            iex_api_hit = api_root + f'/stock/{ticker}/stats'

            r = requests.get(iex_api_hit, params=api_key_payload)
            spy_text_data.append(r.text)
            time.sleep(.1)
```

```python
spy_market_cap_checkpoint = 'spy_stock_data_checkpoint.json'
with open(spy_market_cap_checkpoint, 'w') as f:
    f.write(str(spy_text_data))
    
spy_json_data = []

for text_data in spy_text_data:
    print(text_data[:20])
    if text_data == "Unknown Symbol":
        continue
    basic_stats_object = json.loads(text_data)
    spy_json_data.append(basic_stats_object)
    

stored_symbol_path = 'spy_stock_data.json'
with open(stored_symbol_path, 'w') as f:
    json.dump(spy_json_data, f, indent=2)
```
